```python
e_list_flux = []  # one per flux
key_list = [str(val) for val in np.arange(0, 24, 1)]
for res in res_list:
    params, data, params_list, data_list, branches, dressed_evals, dressed_evecs = res
    insane = np.asarray(data['map_list'][0]).T
    e_dat = {}
    for key, arr in zip(key_list, insane):
        e_vals = (dressed_evals[arr])   # <-- assign to a temporary variable
        e_dat[key] = (e_vals, 'GHz', ['flux', r'$n_r$'])
    e_list_flux.append(e_dat.copy())  # append dict once per flux
```

```python
import numpy as np
import matplotlib.pyplot as plt

def get_intersections(starting_state, flux_arr, e_list_flux):

    # Inputs
    flux_arr = np.array(flux_arr)       # flux sweep
    n_r_list = np.arange(0, 80, 1)       # n_r values
    w_r = 6.627                         # resonator freq in GHz (or whatever)
    n_vals = [1, 2]
    m_vals = [1, 2]
    tol = .0003                          # matching tolerance
    
    pts_flux = []
    pts_n_r = []
    key_list2 = [starting_state]
    
    for f_idx,  in enumerate(flux_arr):
        e_dat = e_list_flux[f_idx]  # dict for this flux point
        for i, key_i in enumerate(key_list2):
            for j, key_j in enumerate(key_list):
                if j <= i:
                    continue
                E_i = np.array(e_dat[key_i][0])  # shape = n_nrs
                E_j = np.array(e_dat[key_j][0])
                _ij = E_j - E_i  # shape = n_nrs
                for n in n_vals:
                    for m in m_vals:
                        ratio = (n*_ij)/(m*w_r)  # shape = n_nrs
                        idx_match = np.where(np.abs(ratio-1)<tol)[0]
                        for idx in idx_match:
                            pts_flux.append()
                            pts_n_r.append(n_r_list[idx])
    return pts_flux, pts_n_r, n_r_list
```

```python
pts_flux0, pts_n_r0, n_r_list = get_intersections('0', flux_arr, e_list_flux)
pts_flux1, pts_n_r1, n_r_list = get_intersections('1', flux_arr, e_list_flux)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def get_intersections_colored(starting_state, flux_arr, e_list_flux, key_list):
    flux_arr = np.array(flux_arr)
    n_r_list = np.arange(0, 80, 1)
    w_r = 6.627
    n_vals = [1]
    m_vals = np.arange(1, 40, 1)
    tol = 0.0002

    intersections = {}  # (i,j): list of (, n_r)

    key_list2 = [starting_state]

    for f_idx,  in enumerate(flux_arr):
        e_dat = e_list_flux[f_idx]
        for key_i in key_list2:
            i = int(key_i)
            for j, key_j in enumerate(key_list):
                if j <= i:
                    continue
                E_i = np.array(e_dat[key_i][0])
                E_j = np.array(e_dat[key_j][0])
                _ij = E_j - E_i
                for n in n_vals:
                    for m in m_vals:
                        ratio = ((_ij) / (m * w_r))
                        idx_match = np.where(np.abs(ratio - 1) < tol)[0]
                        if len(idx_match) > 0:
                            intersections.setdefault((i, j), []).extend(
                                [(, n_r_list[idx]) for idx in idx_match]
                            )

    # --- plotting ---
    import matplotlib as mpl
    cmap = mpl.colormaps.get('tab10')
    
    plt.figure(figsize=(8, 5))
    for k, ((i, j), pts) in enumerate(intersections.items()):
        if not pts:
            continue
        pts = np.array(pts)
        plt.scatter(
            pts[:, 0], pts[:, 1],
            s=.2,
            color=cmap(k / max(1, len(intersections)-1)),
            label=f'{i}->{j}'
        )
    plt.xlabel("Flux (/)")
    plt.ylabel(r"$n_r$")
    plt.legend(title="Transition (ij)", fontsize=8)
    plt.tight_layout()
    #plt.show()

```

```python
get_intersections_colored('0', flux_arr, e_list_flux, key_list)
plt.imshow(mist0, extent=extent, origin='lower', aspect='auto', cmap='magma', vmin=0, vmax=2)
plt.colorbar()
plt.show()
```
