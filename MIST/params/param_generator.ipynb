```python
import itertools
import numpy as np 
```

```python
# Define parameter values
step             = .002
product_baseline = 1
ratio_baseline   = 1
flux_array    = np.arange(-.2, .6, step)
num_points    = len(flux_array)
f_dim         = 30*np.ones(num_points)
r_dim         = 120*np.ones(num_points)
product_array = np.linspace(1, 10, num_points)
ratio_array   = np.linspace(.01, 1, num_points)
```

```python
# Write to text file
with open("param_values_prod.txt", "w") as f:
    for i in range(num_points):
        line = f"{flux_array[i]:.4f}" + " " + f"{product_array[i]:.4f}" + " " + f"{ratio_baseline}"
        f.write(line + "\n")
```

```python
# Write to text file
with open("param_values_ratio.txt", "w") as f:
    for i in range(num_points):
        line = f"{flux_array[i]:.4f}" + " " + f"{product_baseline}" + " " + f"{ratio_array[i]:.4f}"
        f.write(line + "\n")
```

```python
flux_array = np.arange(-0.2, 0.6, 0.002)
product_array = np.linspace(0.5, 2.0, 10)  # EJ*EC values
ratio_baseline = 1.0  # EC/EJ fixed for now

f_dim = 30
r_dim = 120

with open("param_values_prod.txt", "w") as f:
  for prod_val in product_array:
      # Create a clean directory name for this product value
      dir_name = f"prod_{prod_val:.4f}".replace(".", "p")

      for flux in flux_array:
          # Format: flux, product, ratio, f_dim, r_dim, output_dir
          line = f"{flux:.4f} {prod_val:.4f} {ratio_baseline} {dir_name}"
          f.write(line + "\n")

with open("param_values_ratio.txt", "w") as f:
  for ratio_val in ratio_array:
      # Create a clean directory name for this product value
      dir_name = f"ratio_{ratio_val:.4f}".replace(".", "p")

      for flux in flux_array:
          # Format: flux, product, ratio, f_dim, r_dim, output_dir
          line = f"{flux:.4f} {product_baseline} {ratio_val:.4f} {dir_name}"
          f.write(line + "\n")

print(f"Generated {len(flux_array) * len(product_array)} parameter combinations")
```
